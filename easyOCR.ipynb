### Load data

```python
train_df = pd.read_csv('/content/drive/MyDrive/dacon/train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/dacon/test.csv')
submission = pd.read_csv('/content/drive/MyDrive/dacon/sample_submission.csv')
```

```python
from matplotlib.pyplot import imshow
from PIL import Image
import easyocr

test_pred = []

%cd /content/drive/MyDrive/dacon

for i in test_df['img_path']:
   f_name = i
   print(f_name)

   pil_im = Image.open(f_name, 'r')
   cv_img = np.array(pil_im) # numpy로 변환
   gray = cv2.cvtColor(cv_img, cv2.COLOR_BGR2GRAY)  # grayscale
   
  # # easy-ocr
   reader = easyocr.Reader(['ko','en'])
   test_pred.append(reader.readtext(f_name, detail=0))
```

```python
test_pred
```

```python
prediction = []

for i in test_pred:
  i = ''.join(i)
  prediction.append(i)
```

```python
print(prediction)
```

```python
submission['text'] = prediction
```

```python
print(submission)
```
